データ整理後、データ不足分は水増しを行い学習

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense, Dropout, Flatten, Input, BatchNormalization
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import optimizers
from google.colab import files



#データの読込(＋データ水増し)
img_list = ["bookshelf", "chair", "desk", "drawers", "microwaveoven", "refrigerator", "washingmachine"]

img_bookshelf = []
img_chair = []
img_desk = []
img_drawers = []
img_microwaveoven = []
img_refrigerator = []
img_washingmachine = []

furniture_home_app_list = [img_bookshelf, img_chair, img_desk, img_drawers,
                          img_microwaveoven, img_refrigerator, img_washingmachine]

for i in range(len(furniture_home_app_list)):
  path = os.listdir('/content/drive/MyDrive/dataset/'+ img_list[i])

  for j in range(len(path)):
      img = cv2.imread('/content/drive/MyDrive/dataset/' + img_list[i] +"/"+ path[j])
      b,g,r = cv2.split(img)
      img = cv2.merge([r,g,b])
      img = cv2.resize(img, (150,150))
      furniture_home_app_list[i].append(img)
      # 上下反転
      img_ver = cv2.flip(img, 0)
      furniture_home_app_list[i].append(img_ver)
      # 左右反転
      img_hor = cv2.flip(img, 1)
      furniture_home_app_list[i].append(img_hor)
      # 45度回転
      M = cv2.getRotationMatrix2D((img.shape[1]/2, img.shape[0]/2), 45, 1)
      img_rot = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
      furniture_home_app_list[i].append(img_rot)
      # トリミング(画像中心から上下左右に50ピクセルの範囲)
      height = img.shape[0]
      width = img.shape[1]
      center = (int(width/2), int(height/2))
      img_tri = img[center[0] - 50:center[0] + 50, center[1] - 50:center[1] + 50]
      img_tri = cv2.resize(img_tri, (150,150))
      furniture_home_app_list[i].append(img_tri)
      # マスク
      img_rect = cv2.rectangle(img, (30, 50), (120, 100), (0, 0, 255), -1)
      furniture_home_app_list[i].append(img_rect)

X = np.array([x for item in furniture_home_app_list for x in item])
y = np.array([i for i in range(len(furniture_home_app_list)) for _ in range(len(furniture_home_app_list[i]))])

rand_index = np.random.permutation(len(X))
X = X[rand_index]
y = y[rand_index]

# データの分割
X_train = X[:int(len(X)*0.8)]
X_train = X_train[:int(len(X_train)*0.8)]
y_train = y[:int(len(y)*0.8)]
y_train = y_train[:int(len(y_train)*0.8)]
X_val = X[int(len(X)*0.8):]
y_val = y[int(len(y)*0.8):]
X_test = X_train[int(len(X_train)*0.8):]
y_test = y_train[int(len(y_train)*0.8):]

# データ数の確認
print(f"全データ数：{len(X)}")
print(f"トレーニングデータ数：{len(X_train)}")
print(f"バリデーションデータ数：{len(X_val)}")
print(f"テストデータ数：{len(X_test)}")

# 正解ラベルをone-hotに変形
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(y_test)

# モデルにvggを使用
input_tensor = Input(shape=(150, 150, 3))
vgg16 = VGG16(include_top=False, weights='imagenet', input_tensor=input_tensor)

# vggのoutputを受け取り、7クラス分類する層を定義
top_model = Sequential()
top_model.add(Flatten(input_shape=vgg16.output_shape[1:]))
top_model.add(Dense(256, activation='relu'))
top_model.add(Dropout(0.5))
top_model.add(BatchNormalization())
top_model.add(Dense(128, activation='relu'))
top_model.add(Dense(7, activation='softmax'))

# vggと、top_modelを連結
model = Model(inputs=vgg16.input, outputs=top_model(vgg16.output))

# vggの層の重みを固定
for layer in model.layers[:19]:
    layer.trainable = False

# コンパイル
model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.SGD(learning_rate=1e-4, momentum=0.9),
              metrics=['accuracy'])

# 学習
model.fit(X_train, y_train, batch_size=20, epochs=20, verbose=1, validation_data=(X_val, y_val))

# 精度の評価
scores = model.evaluate(X_test, y_test, verbose=1)

print('Test loss:', scores[0])
print('Test accuracy:', scores[1])

# データの可視化（testデータの先頭の10枚）
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.xlabel(f"No.{i + 1}")
    plt.xticks(color="None")
    plt.yticks(color="None")
    plt.imshow(X_test[i].reshape((150, 150, 3)))
plt.suptitle("10 images of test data",fontsize=18)
plt.show()

# 予測（testデータの先頭の10枚）
predction = np.argmax(model.predict(X_test[0:10]), axis=1)
for i, pred in enumerate(predction):
  print(f"{i + 1}番目の予測：", img_list[pred])

model.summary()

# #resultsディレクトリを作成
# result_dir = 'results'
# if not os.path.exists(result_dir):
#     os.mkdir(result_dir)
# # 重みを保存
# model.save(os.path.join(result_dir, 'model.h5'))

# files.download( '/content/results/model.h5' )
```
